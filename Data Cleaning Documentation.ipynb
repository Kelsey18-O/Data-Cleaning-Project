# Data Cleaning Project  Documentation
This notebook documents the process of data cleaning for the customer data dataset. We will inspect the data, handle missing values
and visulize changes before and after cleaning.  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_cleaning.csv.csv")

print(df.info())
print(df.describe())
print(df.head())
```

# Missing Values Before Cleaning 
The following output shows the number of missing values per column.
 

```python
missing_values = df.isnull().sum()
print("Missing Values Per Column:")
print(missing_values)
```

# Data Visualizations Before Cleaning 
These visualizations show the data distributions and missing values before any cleaning is applied.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Before Cleaning')
plt.show()


df.hist(bins=20, figsize=(10, 10), color='skyblue')
plt.suptitle('Data Distributions Before Cleaning')
plt.show()
```

# Data Cleaning Steps
We will now fill missing values and remove duplicates rows to clean the dataset.
 

```python

df.fillna({'Email': "Unknown"}, inplace=True)
df.fillna({'Address': "Unknown"}, inplace=True)
df.fillna({'Avatar': "Unknown"}, inplace=True)
df.fillna({'Avg. Session Length': df['Avg. Session Length'].mean()}, inplace=True)
df.fillna({'Time on App': df['Time on App'].mean()}, inplace=True)
df.fillna({'Time on Website': df['Time on Website'].mean()}, inplace=True)
df.fillna({'Length of Membership': df['Length of Membership'].mean()}, inplace=True)
df.fillna({'Yearly Amount Spent': df['Yearly Amount Spent'].mean()}, inplace=True)

df.drop_duplicates(inplace=True)
```

# Data after cleaning 
We review the data information again to confirm that all the missing values have been addressed. 
 

```python
print("Data Info After Cleaning:")
print(df.info())
print("Missing Values Per Column After Filling:")
print(df.isnull().sum())
```

# Data Visualizations After Cleaning 
The following visualizations show the data distributions and confirm that the missing values have been addressed. 

```python
df.hist(bins=20, figsize=(10, 10), color='lightgreen')
plt.suptitle('Data Distributions After Cleaning')
plt.show()

sns.countplot(x='Avatar', data=df)
plt.xticks(rotation=90)
plt.title('Avatar Count After Cleaning')
plt.show()
```

# Conclusion
In this notebook, we conducted a full data cleaning process:
- Filled missing values in categorical and numeric columns. 
- Removed duplicate rows.
- Obsreved the effects of cleaning through visualizations before and after.

This cleaned data is now ready for further analysis or modeling tasks.
